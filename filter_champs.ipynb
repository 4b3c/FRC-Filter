```python
import json
import pandas as pd

with open("teams_from_champs.json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)
```

```python
grouped = df.groupby("division")

# Example: print all team numbers per division
for division, group in grouped:
    print(f"\nDivision: {division}")
    print(group["team_number"])
```

```python
# Create an Excel writer
with pd.ExcelWriter("teams_by_division.xlsx", engine="xlsxwriter") as writer:
    # Group by division
    for division, group in df.groupby("division"):
        # Clean up sheet name (Excel limits: no special chars, max 31 chars)
        sheet_name = str(division)[:31].replace("/", "_").replace("\\", "_")
        group.to_excel(writer, sheet_name=sheet_name, index=False)

print("Excel file saved as teams_by_division.xlsx")
```
